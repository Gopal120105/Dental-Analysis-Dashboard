```python
import pandas as pd
import numpy as np
import zipfile

zip_path = "3.+Doctors,+Hospitals,+Nurses,+Nursing+Clinics+1+Lac.zip"

with zipfile.ZipFile(zip_path, 'r') as z:
    file_list = z.namelist()

file_list[1:16]
```

```python
excel_files = [
    f for f in file_list
    if f.endswith(".xlsx")
    and "Doctors Dental - 7000" in f
]

len(excel_files)
```

```python
dfs = []
with zipfile.ZipFile(zip_path, 'r') as z:
    for file in excel_files:
        with z.open(file) as f:
            df = pd.read_excel(f)
            df["source_file"] = file.split("/")[-1]  # lineage
            dfs.append(df)
```

```python
dfs
```

```python
dental_df = pd.concat(dfs, ignore_index=True)
```

```python
dental_df.shape
dental_df.head()
```

```python
dental_df.columns = (
    dental_df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
)
```

```python
mobile_cols = [col for col in dental_df.columns if "mobile" in col]

dental_df["mobile_final"] = (
    dental_df[mobile_cols]
    .bfill(axis=1)
    .iloc[:, 0]
)
```

```python
email_cols = [col for col in dental_df.columns if "email" in col]

dental_df["email_final"] = (
    dental_df[email_cols]
    .bfill(axis=1)
    .iloc[:, 0]
)
```

```python
dental_df[["mobile_final", "email_final"]].head(10)
```

```python
dental_df["email_final"] = (
    dental_df["email"]
    .fillna(dental_df.get("email1"))
    .fillna(dental_df.get("companyemail"))
)
```

```python
dental_df["city"] = dental_df["city"].fillna(
    dental_df["source_file"].str.split(" ").str[0]
)
```

```python
dental_df.isnull().sum().sort_values(ascending=False)
```

```python
dental_df.duplicated(
    subset=["company_name", "mobile_final", "city"]
).sum()
```

```python
dental_df.drop_duplicates(
    subset=["company_name", "mobile_final", "city"],
    inplace=True
)
```

```python
dental_df["has_mobile"] = dental_df["mobile_final"].notna().astype(int)
dental_df["has_email"] = dental_df["email_final"].notna().astype(int)
dental_df["has_website"] = dental_df["website"].notna().astype(int)
```

```python
dental_df.to_csv(
    "cleaned_data/dental_dashboard_ready.csv",
    index=False
)
```
